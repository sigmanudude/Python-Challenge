```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "Desktop/homework/matplot/data/city_data.csv"
ride_data_to_load = "Desktop/homework/matplot/data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load, low_memory=False)
ride_df = pd.read_csv(ride_data_to_load, low_memory=False)


# Combine the data into a single dataset
city_ride_df = pd.merge(ride_df, city_df, on="city", how="inner")


# Display the data table for preview
#city_ride_df


#pie plot data
#fare_table for pie
fare_table = city_ride_df.groupby('type')['fare']
fare_table_count = fare_table.count()
fare_table_df = pd.DataFrame(fare_table_count)
#fare_table_df

#total_rides for pie
total_rides = city_ride_df.groupby("type")["driver_count", "city"]
sum_cityride = total_rides.sum()
#sum_cityride

#total_fares for pie
total_fares = city_ride_df.groupby("type")["fare", "city"]
sum_totalfares = total_fares.sum()
#sum_totalfares


#scatter plot data
#city grouped with driver for scatter
city_driver = city_ride_df.groupby('type')['driver_count']
city_driver_count = city_driver.count()
city_driver_df = pd.DataFrame(city_driver_count)

#city groupby test
type_driver = city_ride_df.groupby('type')['driver_count'].mean().sort_values()
type_driver_df = pd.DataFrame(type_driver)

#driver faretest
driver_fare = city_ride_df.groupby('type')['fare'].mean().sort_values()
driver_fare_df = pd.DataFrame(driver_fare)
#driver_fare_df

#city_driver sum
city_driver_sum = city_ride_df.groupby('city')['driver_count'].sum()
city_driver_sum_df = pd.DataFrame(city_driver_sum)

#city_driver_sum

#city fare sum
city_fare_sum = city_ride_df.groupby('city')['fare', 'type'].sum()
city_fare_df = pd.DataFrame(city_fare_sum)

city_fare_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
labels = "Rural", "Suburban", "Urban"
colors = ["gold", "skyblue", "lightcoral"]
# Create a pie chart
plt.scatter(city_driver_sum_df,
            city_fare_df,
            marker="o", 
            facecolors="red", 
            edgecolors="black",
            s=driver_fare_df, 
            alpha=0.50,
            label=labels)
plt.grid()
plt.title("Pybar Ride Sharing Data")
plt.xlabel("Total Number of Rides (per city)")
plt.ylabel("Average Fare")
plt.legend(loc="best")
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "skyblue", "lightcoral"]
# Create a pie chart
plt.pie(
    # using data
    sum_totalfares,
    #labels
    labels=labels,
    # with shadows
    shadow=True,
    # with colors
    colors=colors,
    # with one slide exploded out
    explode=(0, 0, 0.1),
    # with the start angle
    startangle=140,
    # with the percent listed as a fraction
    autopct='%1.1f%%',
    )

plt.show()
```
